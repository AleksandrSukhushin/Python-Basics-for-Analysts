### Задача 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
plt.hist(df['price'], bins = 30, edgecolor = 'k')
plt.xlabel('Цена')
plt.title('Распределение цен на недвижимость')
plt.grid()
plt.show()
```

***Выводы:*** большинство объектов недвижимости имеет цену в диапазоне от 0,1 до 1 миллиона долларов, при этом есть небольшое количество объектов недвижимости с очень высокими ценами.

```python
sns.boxplot(x = df['condition'], y = df['price'], showfliers = False)
plt.xlabel('Состояние дома')
plt.ylabel('Цена')
plt.title('Сравнение цен недвижимости в зависимости от состояния дома')
plt.show()
```

***Вывод:*** состояние дома оказывает значительное влияние на его цену, поэтому дома в лучшем состоянии 5 имеют более высокие цены, чем дома в худшем состоянии 1.

```python
subset = df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'floors', 'condition']]
correlation_matrix = subset.corr()
sns.heatmap(correlation_matrix, annot = True, linewidths = 5, cmap = 'coolwarm')
plt.title('Тепловая карта корреляции')
plt.show()
```

***Вывод:*** цена недвижимости сильно положительно коррелирует с количеством спален и жилой площадью, а также слабо положительно коррелирует с количеством ванных комнат.
