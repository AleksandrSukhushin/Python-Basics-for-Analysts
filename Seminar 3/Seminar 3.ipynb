```python
import pandas as pd
df = pd.read_csv('laptop_price.csv', sep = ',', encoding = 'latin-1')
df.head()
```

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split(' ')[0])
df.head()
```

```python
def convert(x):
    x = x.split(' ')[0]
    if x.endswith('GB'):
        res = x.replace('GB', '')
    elif x.endswith('TB'):
        res = float(x.replace('TB', '')) * 1024
    return int(res)

df['Memory_Amount'] = df['Memory'].apply(convert)
df.head()
```

```python
def convert2(x):
    x = x.split(' ')
    x.pop(0)
    return ' '.join(x)
    # ' '.join(x.split(' ')[1: ])

df['Memory_Type'] = df['Memory'].apply(convert2)
df.head()
```

```python
df_new = df.drop(columns = ['Memory', 'ScreenResolution']) # +inplace = True - сразу удаление
df_new.head(1)
```

```python
df['SSD'] = 0
df.loc[df['Memory_Type'] == 'SSD', 'SSD'] = 1
df.tail(5)
```

```python
df['Weight'] = df['Weight'].str.replace('kg', '').astype('float64')
```
